```python

# ==============================
# Titanic Dataset - Exploratory Data Analysis (EDA)
# Works in Google Colab
# ==============================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Load Titanic dataset
df = sns.load_dataset("titanic")

# ------------------------------
# Basic Information
# ------------------------------
print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

# ------------------------------
# Data Visualization
# ------------------------------

# 1. Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df)
plt.title("25EU02902 Survival Count")
plt.show()

# 2. Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="sex", data=df)
plt.title("25EU02902 Gender Distribution")
plt.show()

# 3. Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=df)
plt.title("25EU02902 Survival by Gender")
plt.show()

# 4. Passenger Class Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="pclass", data=df)
plt.title("25EU02902 Passenger Class Distribution")
plt.show()

# 5. Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x="pclass", hue="survived", data=df)
plt.title("25EU02902 Survival by Passenger Class")
plt.show()

# 6. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["age"].dropna(), bins=30, kde=True)
plt.title("25EU02902 Age Distribution")
plt.xlabel("Age")
plt.show()

# 7. Fare Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["fare"], bins=30, kde=True)
plt.title("25EU02902 Fare Distribution")
plt.xlabel("Fare")
plt.show()

# 8. Age vs Passenger Class
plt.figure(figsize=(8,5))
sns.boxplot(x="pclass", y="age", data=df)
plt.title("25EU02902 Age by Passenger Class")
plt.show()

# 9. Correlation Heatmap
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("25EU02902 Correlation Heatmap")
plt.show()

# 10. Pairplot
sns.pairplot(df[["survived", "age", "fare", "pclass"]].dropna(), hue="survived")
plt.show()

print("\nEDA Completed Successfully!")
```
